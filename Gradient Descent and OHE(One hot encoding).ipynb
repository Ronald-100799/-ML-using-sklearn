## Gradient Descent:

Gradient descent is an optimization algorithm used to find the values of parameters (coefficients) of a function (f) that minimizes a cost function (cost),In other words, it is method used to reduce the error of our ML model iteratively and hence increasing the predictive capability of the model.

```python
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])



def G_D(x,y):
    m_in=0
    b_in=0
    itr=1000
    n=len(x)
    lr=0.08
    for i in range(itr):
        ypred=m_in*x+b_in
        cost=(1/n)*sum([val**2 for val in (y-ypred)])
        md=-(2/n)*sum(x*(y-ypred))
        bd=-(2/n)*sum(y-ypred)
        m_in=m_in-lr*md
        b_in=b_in-lr*bd
        print('m {},b {},cost {} '.format(m_in,b_in,cost))
G_D(x,y)
```

## One Hot encoding to handle categorical(nominal) variables:

Sometimes in datasets, we encounter columns that contain numbers of no specific order of preference. The data in the column usually denotes a category or value of the category and also when the data in the column is label encoded. This confuses the machine learning model, to avoid this the data in the column should be One Hot encoded.

One Hot Encoding 
It refers to splitting the column which contains numerical categorical data to many columns depending on the number of categories present in that column. Each column contains 0 or 1 corresponding to which column it has been placed

```python
price=pd.read_csv(r'D:\BDA 2020 assignments\py-master\ML\5_one_hot_encoding\Exercise\carprices.csv')
```

```python
price.head()
```

```python
price['Car Model'].unique() ##nominal variables
```

```python
dummies=pd.get_dummies(price["Car Model"])
dummies
```

```python
# concat the dataframes
merged=pd.concat([price,dummies],axis='columns')
merged
```

```python
# drop the unecessary columns
final=merged.drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
## bulidinga linear model using sklearn
from sklearn.linear_model import LinearRegression
```

```python
x=final.drop(['Sell Price($)'],axis='columns')
x
```

```python
y=final['Sell Price($)']
y
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
x.columns
```

```python
reg.predict([[4500,4,0,0]]) ## benz ,4yr old,4500 mileage
```

```python
reg.score(x,y) ##acutracy of our model
```

## Tain-Test split practice:

train_test_split is a function in Sklearn model selection for splitting data arrays into two subsets: for training data and for testing data. With this function, you don't need to divide the dataset manually.

By default, Sklearn train_test_split will make random partitions for the two subsets. However, you can also specify a random state for the operation.

```python
## loading the datase
car=pd.read_csv(r'D:\BDA 2020 assignments\py-master\ML\6_train_test_split\carprices.csv')
```

```python
car.shape
```

```python
car.head()
```

```python
## visualize the relationship between variables
plt.scatter(car['Mileage'],car['Sell Price($)']) ##negative linear relationship
```

```python
plt.scatter(car['Age(yrs)'],car['Sell Price($)']) ## again a negative trend
```

```python
X=car[['Mileage','Age(yrs)']]
X
```

```python
Y=car['Sell Price($)']
Y
```

```python
## splitting the dataset into test and train 
from sklearn.model_selection import train_test_split

X_Train,x_test,Y_Train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

clf=LinearRegression()
clf.fit(X_Train,Y_Train)
```

```python
clf.predict(x_test)
```

```python
y_test
```

```python
clf.score(X_Train,Y_Train)
```
